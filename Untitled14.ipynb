```python
# Function to load images from a folder
def load_images_from_folder(folder):
    images = []
    filenames = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, filename))
        if img is not None:
            images.append(img)
            filenames.append(filename)
    return images, filenames
```

```python
# Folder path
images_from_folder = "C:/Users/deeph/Desktop/New folder (2)"
images, filenames = load_images_from_folder(images_from_folder)
print(f"Number of images loaded: {len(images)}")
```

```python
# Perform detection
detection_output = model.predict(images, conf=0.25, save=False)
```

```python
# Directory to save images with detections
output_dir = "C:/Users/deeph/Desktop/Detected Images"
os.makedirs(output_dir, exist_ok=True)
```
